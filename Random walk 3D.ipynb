```python
#python code for Random walk in 3-Dimension.
import random 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
n = 100
```

```python
x = np.zeros(n)
y = np.zeros(n)
z = np.zeros(n)
```

```python
for i in range(1, n):
    val = random.randint(1,6)
    if val==1:
        x[i] = x[i-1] + 1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif val==2:
        x[i] = x[i-1] - 1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif val==3:
        x[i] = x[i-1] 
        y[i] = y[i-1] + 1
        z[i] = z[i-1]
    elif val==4:
        x[i] = x[i-1] 
        y[i] = y[i-1] - 1
        z[i] = z[i-1]
    elif val==5:
        x[i] = x[i-1] 
        y[i] = y[i-1] 
        z[i] = z[i-1] + 1
    else:
        x[i] = x[i-1] 
        y[i] = y[i-1] 
        z[i] = z[i-1] - 1
        
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'black')
```
